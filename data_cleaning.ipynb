```python
#import pandas library
import pandas as pd
import datetime as dt
```

```python
#Inserting website url
url= 'https://konbriefing.com/en-topics/cyber-attacks-2022.html'
```

```python
#Getting the first element of dataframe from the list
df= pd.read_html(url)[0]
```

# 1. Drop unwanted columns

```python
#Drop serial number column
df= df.drop(['#'], axis= 1)
```

```python
#Rename column
df.rename(columns={'Stichworte': 'Keywords'}, inplace= True)
```

# 2. Null Values

```python
#Checking for null values
df.isnull().sum()
```

```python
#Percentage of null values in columns
df.isnull().mean()*100
```

```python
#Fetching records where Country value is missing
df[df['Country'].isnull()]
```

```python
#Checking for the presence of values in City column where values in Country column is missing
df[df['Country'].isnull()]['City'].isnull().sum()
```

Since the number is same, so there is no records of City available where values for Country is missing, hence we drop all such records.

##2.1 Country

```python
#Dropping all index values where values in Country are null
df.drop(index= df[df['Country'].isnull()].index, inplace= True, axis=0)

#or

#df= df.dropna(subset=['Country'])
```

```python
df.isnull().sum()
```

## 2.2 State

```python
#Fetching records where State values are missing
df[df['State'].isnull()]
```

```python
#Checking for the presence of City name in records where Values for State are missing
df[df['State'].isnull()]['City'].isnull().sum()
```

Since there are 600(801-201) records in City available, where values in State are missing. So we will fill up the State with consequent city name, by assuming here that City and State names are same.

```python
#Filling the missing values of State using filling forward from City column
df[['City', 'State']] = df[['City', 'State']].fillna(method='ffill', axis=1, inplace= False)
```

```python
df.isnull().sum()
```

```python
#Fetching recors where values for State are missing
df[df['State'].isnull()]
```

## 2.3 City, State and Type

```python
#Filling null values in City and State columns with "Missing"
df['City'].fillna('missing', axis=0, inplace= True)
df['State'].fillna('missing', axis=0, inplace= True)
df['Type'].fillna('In process of analysis', axis=0, inplace= True)
```

```python
df.isnull().sum()
```

# 3. Correct values in Date column

```python
df['Date'].head(30)
```

There are some records in Date column, which are missing with either month or year information. So we can use fill forward method for Date column because of a specific order, it will help with further analysis.

```python
#Convert the 'Date' column to datetime format
df['Date']= pd.to_datetime(df['Date'], errors= 'coerce')
```

```python
#Filling the missing values of month or the date by using filling forward from record above
df['Date'] = df['Date'].fillna(method= 'ffill')
```

```python
df['Date'].head(20)
```

```python
df.isnull().sum()
```
